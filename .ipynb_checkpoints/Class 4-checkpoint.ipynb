```python
#singly linked list

class SinglyLinkedList:
    class Node:
        def __init__(self, data):
            self.data = data
            self.next = None
    def __init__(self):
        self.head = None
    def append(self, value):
        new_node = self.Node(value)
        if self.head == None:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            
        #CLASS CHAL***
    def insert(self, index, value):
        #what if the list is empty?***
        #check to see if self.head is None and if not self.head I can raise error.
        #what if index is 0?***
        #I can have index == 0 and set newnodenext = self.head and self.head = newnode
        #what if there is only 1 element***
        #what if there is more than 1 element***
        #current can be self.head
        #what if the index exceeds the length of the list?***
        #use a range from 0 to the index and raise an error if ever not current
        #now past all the special cases, have current = current.next to get to the next node
        #now add the new node so newnodenext = currentnext
        #currentnext = newnode
        #CLASS CHAL***
        
        #teacher psudocode for insert*****
        #define function for insert with paramaters index and value
        #create a new node with the provided value:
        #if the list is empty aka not head
            #then, simply update our new node next reference so it is pointed at the head
            #and update the head reference so that it points to the new node
        #otherwise
            #we need to validate that the index within our list
            #create a variable to keep track of how many times we traverse
            #count = 0
            #current is set to head
            #create a reference to the previous node(previous) which starts as None
            #while count does not equal index and there is a next node to traverse into
                #update previous to the same node as current
                #current is updated to the next node in the sequence
                #add 1 to count
            #if previous was not updated, then, it means there was only one node in the list
                #append the newly created node to my list
            #otherwise if count is equal to the given index:
            #this means we were able to find the index in the scope of the list
            #previous next reference is updated to point to the newly created node
            #our newly created node's next reference is updated to point to current
                #otherwise
                #this case happens when index is > count
                #index fell outside the scope of the list
                #append the newly created node to my list
        #teacher code end ******
        new_node = self.Node(value)
        if self.head == None:
            if index == 0:
                self.head = new_node
            else:
                raise ValueError("Index out of range")
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            prev = None
            count = 0
            while count < index and current:
                prev = current
                current = current.next
                count += 1
            if count != index:
                raise ValueError("Index out of range")
            prev.next = new_node
            new_node.next = current
    def __str__(self):
        out = "["
        if self.head:
            out += "%s" % self.head.data
            current = self.head.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
```

```python
#doubly linked lists

class DoublyLinkedList:
    class Node:
        def __init__(self,data):
            self.data = data
            self.next = None
            self.prev = None
            
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
    
    def append(self, value):
        new_node = self.Node(value)
        if self.tail:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.length += 1
        if not self.head:
            self.head = new_node
            self.tail = new_node
            self.length += 1
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node
            new_node.prev = current
            self.length += 1
            
    def search(self, value):
        if not self.head:
            return False
        else:
            current = self.head
            tail = self.tail
            while current != tail:
                #now to traverse
                if current.data != value and tail.data != value:
                    current = current.next
                    tail = tail.prev
                else:
                    return True
            return False
        
        #implement binary search as a method to this class (DLL)
    def binary_search(self, value):
        #psudo can be found on wiki
        if not self.head:
            return False
        else:
            left = self.head
            right = self.tail
            while left != right:
                mid = left
                count = 0
            while mid != right and count < (self.length // 2):
                mid = mid.next
                count += 1
            if mid.data == value:
                return True
            elif mid.data > value:
                right = mid.prev
            else:
                left = mid.next
        return left.data == value
            
    def __str__(self):
        out = "["
        if self.head:
            out += "%s" % self.head.data
            current = self.head.next
            while current:               #while current != None aka while current has stuff
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out
```

```python
dll = DoublyLinkedList()

print(dll)

dll.append(1)

print(dll)

rangee = 11
for number in range(2,rangee):
    dll.append(number)

print(dll)
```
